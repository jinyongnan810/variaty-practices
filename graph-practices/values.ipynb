# Plot values

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Plotting values with arrows

```python
_, ax=plt.subplots(figsize=(18, 6))
ax.set_title('Average Monthly Temperature in New York City')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (F)')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax.set_yticks(range(0, 101, 10))
ax.set_yticklabels(range(0, 101, 10))
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.set_xlim(1, 12)
ax.set_ylim(0, 100)
ax.fill_between(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35], 
                color='skyblue', alpha=0.4)
ax.plot(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35],
         color='Slateblue', alpha=0.6, linewidth=2)
ax.annotate('Peak Temperature', xy=(7, 85), xytext=(8, 90),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Lowest Temperature', xy=(1, 30), xytext=(2, 20),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

## Plotting two difference kind of values in a same graph

```python
# Plotting two difference kind of values in a same graph
_, ax = plt.subplots(figsize=(18, 6))
ax.set_title('Average Monthly Temperature and Precipitation in New York City')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (F) / Precipitation (inches)')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax.set_yticks(range(0, 101, 10))
ax.set_yticklabels(range(0, 101, 10))
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.set_xlim(1, 12)
ax.set_ylim(0, 100)
ax.fill_between(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35], 
                color='skyblue', alpha=0.4)
ax.plot(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35],
         color='Slateblue', alpha=0.6, linewidth=2, label='Temperature (F)')
ax2 = ax.twinx()
ax2.set_ylabel('Precipitation (inches)')
ax2.set_ylim(0, 10)
ax2.set_yticks(range(0, 11, 1))
ax2.set_yticklabels(range(0, 11, 1))    
ax2.fill_between(range(1, 13), [3, 2, 4, 5, 6, 7, 8, 9, 8, 7, 5, 4], 
                color='orange', alpha=0.4)
ax2.plot(range(1, 13), [3, 2, 4, 5, 6, 7, 8, 9, 8, 7, 5, 4],
            color='orange', alpha=0.6, linewidth=2, label='Precipitation (inches)')
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```
