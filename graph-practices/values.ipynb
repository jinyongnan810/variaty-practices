# Plot values

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
_, ax=plt.subplots(figsize=(18, 6))
ax.set_title('Average Monthly Temperature in New York City')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature (°F)')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
ax.set_yticks(range(0, 101, 10))
ax.set_yticklabels(range(0, 101, 10))
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.set_xlim(1, 12)
ax.set_ylim(0, 100)
ax.fill_between(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35], 
                color='skyblue', alpha=0.4)
ax.plot(range(1, 13), [30, 32, 45, 60, 70, 80, 85, 84, 75, 65, 50, 35],
         color='Slateblue', alpha=0.6, linewidth=2)
ax.annotate('Peak Temperature', xy=(7, 85), xytext=(8, 90),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Lowest Temperature', xy=(1, 30), xytext=(2, 20),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
