# Histgram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 100 random numbers
df = pd.DataFrame({
    'id': range(100),
    'value': np.random.uniform(low=0, high=100, size=100)
}) 
_, ax=plt.subplots(figsize=(18, 6))
# custom bins
bins = np.linspace(0, 100, 21)
values, bins, patches = ax.hist(df['value'], bins=bins, color='skyblue', alpha=0.7)
ax.set_title('Histogram of Random Numbers')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_xticks(bins)
ax.set_xticklabels([f'{int(b)}' for b in bins], rotation=45)

# show value at the top of each bar
for i in range(len(bins)-1):
    count = values[i]
    ax.text((bins[i] + bins[i+1]) / 2, count, str(count), ha='center', va='bottom')

# add border to each bar
for i in range(len(bins)-1):
    count = values[i]
    ax.add_patch(plt.Rectangle((bins[i], 0), bins[i+1]-bins[i], count, edgecolor='black', facecolor='none', lw=1))

# print values
for i in range(len(bins)-1):
    count = values[i]
    print(f'Bin {i}: {bins[i]}-{bins[i+1]}: {count}')

# show arrows at highest and lowest values
for i in values:
    if i == max(values):
        x = bins[values.tolist().index(i)] + (bins[values.tolist().index(i)+1] - bins[values.tolist().index(i)]) / 2
        ax.annotate('Max', xy=(x, i), xytext=(x+1, i+1),
                    arrowprops=dict(facecolor='orange', shrink=0.05))
    if i == min(values):
        x = bins[values.tolist().index(i)] + (bins[values.tolist().index(i)+1] - bins[values.tolist().index(i)]) / 2
        ax.annotate('Min', xy=(x, i), xytext=(x-1, i-1),
                    arrowprops=dict(facecolor='orange', shrink=0.05))

plt.show()
```
